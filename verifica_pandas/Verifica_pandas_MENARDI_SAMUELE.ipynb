```python
data = pd.read_csv("world_economic_indicators.csv")
```

```python
data
```

```python
#Scartiamo le colonne futili:
data_filtered = data[["Country Name", "Year", "Unemployment, total (% of total labor force)", "GDP (current US$)_x", "GDP growth (annual %)_x"]]
```

```python
data_filtered
```

```python
#Punto 1:
#Stampa la lista dei paesi contenuti nel file.
states = data_filtered[data_filtered["Year"] == 1960]["Country Name"].values
states #Array degli stati
```

```python
#Punto 2:
#Calcola il prodotto interno lordo del mondo (ottenuto sommando quello di tutti i paesi) per ogni anno tra il 1960 e il 2022.
data_filtered[["Year", "GDP (current US$)_x"]].groupby("Year").sum()
```

```python
#Punto 3:
#Calcola il valore massimo della crescita percentuale del prodotto interno lordo per gli Stati Uniti.
data_filtered[data["Country Name"] == "United States"]["GDP growth (annual %)_x"].max()
```

```python
#Punto 4:
#Estrai la percentuale di disocupazione in Italia nel 2022.
data_filtered[(data_filtered["Country Name"] == "Italy") & (data_filtered["Year"] == 2022)]["Unemployment, total (% of total labor force)"]
```

```python
#Punto 5:
#Estrai il prodotto interno lordo dell'Italia in tutti gli anni tra il 1960 e il 2022.
italy_2022_GDP = data_filtered[(data_filtered["Country Name"] == "Italy")][["Year", "GDP (current US$)_x"]]
italy_2022_GDP
```

```python
#Punto 6:
#Con i dati del punto 5) crea un grafico a linea che rappresenti l'andamento nel tempo del prodotto interno lordo dell'Italia.
#Il grafico deve avere etichette degli assi corrette, titolo e griglia.
fig, ax = plt.subplots(figsize = (4, 4))

time = italy_2022_GDP["Year"].values #Sull'asse x
gdp = italy_2022_GDP["GDP (current US$)_x"].values #Sull'asse y

ax.plot(time, gdp)

ax.set_title("L'andamento nel tempo del prodotto interno lordo dell'Italia")
ax.set_xlabel("Year")
ax.set_ylabel("Prodotto Interno Lordo")

ax.grid()
plt.show()
```

```python
#Punto 7:
#Crea un grafico con due linee che rappresentino l'andamento nel tempo del prodotto interno lordo dell'Italia e della Francia.
#Il grafico deve avere etichette degli assi corrette, titolo, griglia e legenda.
france_2022_GDP = data_filtered[(data_filtered["Country Name"] == "France")][["Year", "GDP (current US$)_x"]]

fig, ax = plt.subplots(figsize = (10, 4))

time = italy_2022_GDP["Year"].values #Sull'asse x
gdp_italy = italy_2022_GDP["GDP (current US$)_x"].values #Sull'asse y
gdp_france = france_2022_GDP["GDP (current US$)_x"].values

ax.plot(time, gdp_italy, label = "PIL Italia", c = "red")
ax.plot(time, gdp_france, label = "PIL Francia", c = "blue")

ax.set_title("Confronto tra il PIL italiano e il Pil francese")
ax.set_xlabel("Year")
ax.set_ylabel("Prodotto Interno Lordo")

ax.grid()
ax.legend()
plt.show()
```

```python
#Punto 8:
#Elimina tutti i dati mancanti.
#Traccia un grafico a dispersione che permetta di rispondere alla domanda seguente: 
#è vero che i paesi con alto tasso di disoccupazione hanno un prodotto interno lordo basso?
data_without_Nan = data_filtered.copy()

data_without_Nan.dropna(inplace = True)
data_without_Nan

unemployment = data_without_Nan["Unemployment, total (% of total labor force)"].values
gdp = data_without_Nan["GDP (current US$)_x"].values

fig, ax = plt.subplots(figsize = (10, 3))

ax.scatter(unemployment, gdp, alpha = 0.5, s = 8)

ax.set_xlabel("unemployment")
ax.set_ylabel("PIL")
ax.set_title("Confronto tra il PIL e il numero di disoccupati")

plt.show()

#La risposta alla domanda è corretta, all'aumentare del numero di disoccupati, il PIL diminuisce
#Quando il numero di disoccupati è basso, il PIL è alto.
```

# Esercizio 2: Pokemon
Il file `pokemon.csv` contiene una lista di pokemon, ciascuno con il proprio tipo (Type1 e Type2). La cartella `./images` contiene le immagini dei pokemon in formato .png. Il nome file corrisponde al nome del pokemon.

Rispondi alle seguenti domande utilizzando pandas, matplotlib ed eventualmente numpy.

1) Quanti sono i pokemon per ciascun possibile valore di tipo 1 `Type1`?

2) Crea una figura in cui riporti un pokemon casuale per ciascuno valore di tipo 1 `Type1`. Struttura la figura su più righe e colonne. Elimina gli assi cartesiani e riporta il nome di ciascun pokemon nel grafico.

```python
# soluzione qui
data = pd.read_csv("pokemon.csv")
data
```

```python
#Punto 1:
#Quanti sono i pokemon per ciascun possibile valore di tipo 1 Type1?
data["Type1"].value_counts()
```

```python
#Punto 2:
#Crea una figura in cui riporti un pokemon casuale per ciascuno valore di tipo 1 Type1.
#Struttura la figura su più righe e colonne.
#Elimina gli assi cartesiani e riporta il nome di ciascun pokemon nel grafico.

dim = len(data["Type1"].value_counts()) #Numero di possibili tipi di Type1
types = data["Type1"].values #Lista di tutti i valori di Type1
possible_types = np.random.choice(types, dim, replace = False) #Valori univoci del tipo Type1

#Strutturo la figura come una matrice
fig, axs = plt.subplots(nrows = 2, ncols = dim // 2, figsize = (12, 4))

index = 0 #Indice utile per scorrere l'array di tutti i possibili tipi univoci
for riga in range(0, 2):
    for colonna in range(0, dim // 2):
        #Pokemon è scelto tra tutti i nomi possibili dei 18 possibili tipi di Type1
        pokemon = np.random.choice((data[data["Type1"] == possible_types[index]]["Name"].values))
        
        photo = plt.imread(f"./images/{pokemon}.png") #Lettura della foto
        
        axs[riga][colonna].imshow(photo)
        axs[riga][colonna].set_title(pokemon)
        
        #Si tolgono gli assi
        axs[riga][colonna].axis("off")
        
        index += 1
        
plt.suptitle("POKEMON")
plt.show()
```
