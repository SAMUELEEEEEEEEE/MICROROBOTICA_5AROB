### Esercizio su _sei personaggi in cerca di autore_ di Pirandello
Descrizione esercizio :
- **Quante volte una parola è ripetuta**

```python
def read_file(nomeFile : str):
    """
    This function returns the list of words to count
    Parameters:
    -nomeFile: str, file name
    
    Returns:
    -parole: list of string
    """
    with open (nomeFile, "r", encoding = "utf-8") as file:
        readFile = file.readlines()
    
    file_senza_punteggiatura = ""#File senza punteggiatura alcuna
    punteggiatura = (".", ";", ",", ":", "(", ")", "?", "!", "'\'", "\"", "\n", "\t", "'", "<<", ">>", "-")#tupla di caratteri di punteggiatura
    accentate = {"à" : "a", "ò" : "o", "è" : "e", "ù" : "u", "ì" : "i"}#dizionario di lettere accentate
    
    for riga in readFile:
        for parola in riga:
            parola_senza_accenti = ""
            for lettera in parola:
                if lettera.lower() in accentate: #Metodo per togliere le lettere accentate sostituendole con lettere senza accento
                    parola_senza_accenti += accentate[lettera.lower()]
                else:
                    parola_senza_accenti += lettera.lower()
                    
            if parola_senza_accenti not in punteggiatura:
                file_senza_punteggiatura += parola_senza_accenti.lower()
            else:
                file_senza_punteggiatura += " "
                
                
    parole = file_senza_punteggiatura.split(" ")
    parole = [parola for parola in parole if len(parola)>0]#Lista di parole senza spazi vuoti
            
    return parole
```

```python
def conteggio_parole(parole : list):
    dizionario_parole_ripetizione = {} #{"parola" : numeroRipetizioneZ}
    for parola in parole:
        if parola not in dizionario_parole_ripetizione:
            dizionario_parole_ripetizione[parola] = 1
        else:
            dizionario_parole_ripetizione[parola] += 1
    return dizionario_parole_ripetizione
```

```python
parole = read_file("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt")
conteggio_parole(parole)
```
