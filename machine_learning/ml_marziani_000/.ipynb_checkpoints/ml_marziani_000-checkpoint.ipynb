```python
colori = np.sort(data["colore"].dropna().unique())
```

```python
colori
```

```python
data_modified = data.copy()
```

```python
for k in range(len(colori)):
    data_modified["colore"].replace(colori[k], k, inplace = True)
```

```python
data_modified
```

```python
import seaborn as sns
```

```python
sns.set_theme(font_scale = 2)
sns.pairplot(data_modified, hue = "specie", dropna = True)
```
