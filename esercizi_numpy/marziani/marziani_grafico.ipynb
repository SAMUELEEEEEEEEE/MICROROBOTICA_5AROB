```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("marziani.csv")
```

```python
data
```

```python
data.dropna(inplace = True)
```

```python
caratteristiche = ("colore", "n_arti", "peso", "altezza", "larghezza")
```

```python
dim = len(caratteristiche)
```

```python
fig, axs = plt.subplots(ncols = dim, nrows = dim, figsize = (16, 16), constrained_layout = True)
for indice_riga, caratteristica_riga in enumerate(caratteristiche):
    for indice_colonna, caratteristica_colonna in enumerate(caratteristiche):
        if indice_colonna != indice_riga:
            axs[indice_riga][indice_colonna].scatter(data[data["specie"] == "Simmy"][f"{caratteristica_colonna}"].values,
                                                    data[data["specie"] == "Simmy"][f"{caratteristica_riga}"].values, 
                                                     c = "r", label = "Simmy")
            
            axs[indice_riga][indice_colonna].scatter(data[data["specie"] == "Robby"][f"{caratteristica_colonna}"].values,
                                                    data[data["specie"] == "Robby"][f"{caratteristica_riga}"].values, 
                                                     c = "g", label = "Robby")
            axs[indice_riga][indice_colonna].set_xlabel(f"{caratteristica_colonna}")
            axs[indice_riga][indice_colonna].set_ylabel(f"{caratteristica_riga}")
        else:
            axs[indice_riga][indice_colonna].hist(data[data["specie"] == "Simmy"][f"{caratteristica_colonna}"].values, 
                                                     color = "r", label = "Simmy", alpha = 0.5, bins = 10)
            
            axs[indice_riga][indice_colonna].hist(data[data["specie"] == "Robby"][f"{caratteristica_colonna}"].values, 
                                                     color = "g", label = "Robby", alpha = 0.5, bins = 10)
            axs[indice_riga][indice_colonna].set_xlabel(f"{caratteristica_colonna}")
            axs[indice_riga][indice_colonna].set_ylabel("NÂ° di esemplari")
            
            
fig.suptitle("Le caratteristiche dei marziani", fontsize = 25)
fig.legend(labels = ["Simmy", "Robby"], ncols = 2, fontsize = 18)
plt.show()
```
