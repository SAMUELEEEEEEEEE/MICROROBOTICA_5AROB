```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_json("neowise.json")
```

```python
data
```

```python
data[["orbit_class", "period_yr"]].groupby("orbit_class").mean()
```

```python
filtered = data[data["period_yr"] < 40]
```

```python
period_yr = filtered["period_yr"].values
moid_au = filtered["moid_au"].values
```

```python
fig, ax = plt.subplots(figsize = (3, 3))
ax.scatter(moid_au, period_yr, s = 8, alpha = 0.5)

ax.set_xlabel("moid_au")
ax.set_ylabel("period_yr")
ax.set_title("asteroidi & comete")
ax.grid()

plt.show()
```
