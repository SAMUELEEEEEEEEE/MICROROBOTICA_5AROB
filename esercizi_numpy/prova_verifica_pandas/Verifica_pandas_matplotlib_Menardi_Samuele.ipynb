Il file `Netflix-Dataset-Latest-2021.xlsx` contiene svariati dati di film e serie Netflix.

1) Tra tutti film/serie che durano meno di 30 minuti calcola il valore minimo, massimo e medio dello score (`IMDb Score`).

2) Calcola, mediante una sola riga di codice, quanti sono i film (Movie) e le serie (Series) presenti nel file (`Series or Movie`).

3) In media hanno uno score maggiore (`IMDb Score`) i film/serie che durano meno di 30 minuti oppure quelli che durano pi√π di 2h?

4) Utilizza matplotlib creare un grafico a dispersione che mostri la correlazione tra `IMDb Score` e `Awards Received`. I punti del grafico devono avere un colore che dipenda dal numero di voti ricevuti `IMDb Votes`. I grafici devono avere titolo, griglia ed etichette sugli assi.

5) Traccia lo stesso grafico facendo in modo che la dimensione dei punti sia proporzionale al numero di voti `IMDb Votes`.

6) Traccia in un'unica figura due due istogrammi che rappresentino rispettivamente le distribuzioni di `IMDb Score` e `Awards Received`.

```python
#soluzione qui
```

```python
data = pd.read_excel("Netflix-Dataset-Latest-2021.xlsx")
```

```python
#Risoluzione punto 1:
data[data["Runtime"] == "< 30 minutes"]["IMDb Score"].min()
```

```python
data[data["Runtime"] == "< 30 minutes"]["IMDb Score"].max()
```

```python
data[data["Runtime"] == "< 30 minutes"]["IMDb Score"].mean()
```

```python
#Risoluzione punto 2:
data["Series or Movie"].value_counts()
```

```python
#Risoluzione punto 3:
score_min_30 = data[data["Runtime"] == "< 30 minutes"]["IMDb Score"].mean()
```

```python
score_2_h = data[data["Runtime"] == "> 2 hrs"]["IMDb Score"].mean()
```

```python
score_min_30 > score_2_h
#E' maggiore lo score dei film/serie con una durata minore di 30 minuti
```

```python
#Risoluzione punto 4:
imdb_score = data["IMDb Score"].values
awards_received = data["Awards Received"].values
imdb_votes = data["IMDb Votes"].values
```

```python
fig, ax = plt.subplots(figsize = (5, 5))

ax.scatter(imdb_score, awards_received, c = imdb_votes)

ax.set_xlabel("IMDb Score")
ax.set_ylabel("Awards Received")

ax.grid()
ax.set_title("Awards & Score with different shapes")

plt.show()
```

```python
#Risoluzione punto 5:
fig, ax = plt.subplots(figsize = (5, 5))

ax.scatter(imdb_score, awards_received, c = imdb_votes, s = imdb_votes / len(imdb_score))

ax.set_xlabel("IMDb Score")
ax.set_ylabel("Awards Received")

ax.grid()
ax.set_title("Awards & Score")

plt.show()
```

```python
#Risoluzione punto 6:
fig, axs = plt.subplots(ncols = 2, figsize = (12, 5))

axs[0].hist(imdb_score, bins = 50)
axs[0].set_xlabel("IMDb Score")
axs[0].set_ylabel("Frequency")
axs[0].grid()
axs[0].set_title("IMDb Score")

axs[1].hist(awards_received, bins = 50)
axs[1].set_xlabel("Awards Received")
axs[1].set_ylabel("Frequency")
axs[1].grid()
axs[1].set_title("Awards Received")

plt.suptitle("IMDb Score & Awards Received")

plt.show()
```
