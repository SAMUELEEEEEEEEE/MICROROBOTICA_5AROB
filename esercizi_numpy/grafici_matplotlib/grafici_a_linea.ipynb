# Esercizio
funzioni:

X ** 2

X ** 3

X ** 4

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 4000)
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4
```

```python
fig, ax = plt.subplots(figsize = (6, 6))
ax.plot(x, y1, "r", linewidth = "2", label = "$x^2$")
ax.plot(x, y2, "b", linewidth = "4", label = "$x^3$")
ax.plot(x, y3, "g", linewidth = "6", label = "$x^4$")

ax.grid()

ax.set_title("Grafico di potenze di x")
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.legend()

plt.savefig("grafico.pdf", dpi = 200)

plt.show()
```
