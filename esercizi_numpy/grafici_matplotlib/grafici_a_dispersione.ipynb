## Grafici a dispersione
Serve per comprendere se x dipende da y

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
x = np.random.normal(0., 1., n)
y = x + np.random.normal(2, 1., n)
colors = np.random.choice(["r", "g", "b", "k", "y"], n)
sizes = np.random.uniform(1, 50, n)
```

```python
fig, ax = plt.subplots(figsize = (3, 3))
#.scatter(x, y) per generare un grafico a dispersione
#.scatter puo personalizzare ogni singolo punto
ax.scatter(x, y, c = colors, s = sizes, marker = "*")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Grafico a dispersione")

plt.show()
```
