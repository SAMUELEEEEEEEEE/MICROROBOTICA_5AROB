```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#Usando matplotlib con approccio OO (Object Oriented), opposto al pyplot
fig, ax = plt.subplots(figsize = (6 , 3))
#fig è la figura
#ax è un oggetto Axes(1 Axes = 1 Grafico)
ax.plot(x, y1, "k-.", linewidth = 2, label = "sin(x)")
ax.plot(x, y2, "r", linewidth = 4, label = "cos(x)")#Per aggiungere una linea
ax.set_title("Un grafico a linea", fontsize = 12)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()  #Creare una legenda
ax.grid()  #griglia
plt.show()
```
