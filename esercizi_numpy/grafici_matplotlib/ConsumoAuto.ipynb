```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file=open("Fuel_Consumption_2000-2022.csv","r")
lines=file.readlines()
file.close()
linee_senza_a_capo=[line[:-1] for line in lines if len(line)>1]
```

```python
fuel_consumptions=[]
make=[]
emissions=[]
for riga in linee_senza_a_capo[1:]:
    lista=riga.split(",")
    fuel_consumptions.append(float(lista[8]))
    emissions.append(int(lista[12]))
    make.append(lista[1])
fuel_consumptions=np.array(fuel_consumptions)
emissions=np.array(emissions)
make=np.array(make)
```

```python
fig, ax=plt.subplots(figsize=(3,3)) 
ax.scatter(fuel_consumptions,emissions, c="k", s=5, alpha=0.5)
ax.set_xlabel("fuel_consumptions")
ax.set_ylabel("emissions")

plt.show()
```

```python
diz_marche_occorrenze = {}
for riga in linee_senza_a_capo[1:]:
    lista = riga.split(",")
    if lista[1] in diz_marche_occorrenze:
        diz_marche_occorrenze[lista[1]] += 1
    else:
        diz_marche_occorrenze[lista[1]] = 1
diz_marche_occorrenze
```

```python
listaIndiciValori = list(diz_marche_occorrenze.items()) # lista di tuple(indice, valore)
listaIndiciValori.sort(key=lambda x: x[1], reverse=True) # lista decrescente ordinata per Valore
```

```python
marche=[m[0] for m in listaIndiciValori[:5]]
marche
```

```python
fig, ax=plt.subplots(figsize=(5,5))
for marca in marche:
    arrayBool=make==marca
    c=np.random.uniform(0,1,3)
    ax.scatter(fuel_consumptions[arrayBool],emissions[arrayBool], c=c, s=5, alpha=0.5, label=marca)
ax.set_xlabel("fuel_consumptions")
ax.set_ylabel("emissions")
ax.legend()
plt.show()
```
