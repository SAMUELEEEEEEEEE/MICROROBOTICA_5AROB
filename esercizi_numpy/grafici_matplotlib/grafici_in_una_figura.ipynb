```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Fuel_Consumption_2000-2022.csv")
data
```

```python
engine_size = data["ENGINE SIZE"].values
```

```python
cylinders = data["CYLINDERS"].values
```

```python
fuel_consumption = data["FUEL CONSUMPTION"].values
```

```python
emissions = data["EMISSIONS"].values
```

```python
coppie = (
    (engine_size, cylinders, "engine_size", "cylinders"),
    (engine_size, fuel_consumption, "engine_size", "fuel_consumption"),
    (engine_size, emissions, "engine_size", "emissions"),
    (cylinders, fuel_consumption, "cylinders", "fuel_consumption"),
    (cylinders, emissions, "cylinders", "emissions"),
    (fuel_consumption, emissions, "fuel_consumption", "emissions")
    )
```

```python
fig, axs = plt.subplots(nrows = 6, figsize = (10, 20))
axs[0].set_title("Grafici")
for index, coppia in enumerate(coppie):
    axs[index].scatter(coppia[0], coppia[1], c = "b", s = 10, alpha = 0.5)
    axs[index].set_xlabel(coppia[2])
    axs[index].set_ylabel(coppia[3])
    axs[index].grid()
```
