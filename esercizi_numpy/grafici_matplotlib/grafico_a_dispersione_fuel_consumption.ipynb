# Grafico a dispersione per il consumo del carburante

##### A) x = fuel consumption ; y = emission 
##### B) Cinque marche che consumano di pi√π, per ogni marca un colore

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Indici nel csv per gli array
#fuel consumption = [8]
#emission = [-1]

fileName = "Fuel_Consumption_2000-2022.csv"

with open(fileName, "r", encoding = "utf-8") as file:
    string_file = file.read()
string_list = string_file.split("\n")
```

```python
#Si parte da string_list[1] per scartare il tracciato record 
#fino a -1 per evitare l'ultima riga bianca

fuel_consumption_list = [float(line.split(",")[8]) for line in string_list[1 : -1]]
emission_list = [float(line.split(",")[-1]) for line in string_list[1 : -1]]
make_list = [line.split(",")[1] for line in string_list[1 : -1]]

fuel_consumption_array = np.array(fuel_consumption_list)
emission_array = np.array(emission_list)
make_array = np.array(make_list)
```

```python
#Risoluzione del punto A) -> x e y dipendono -> al crescere di x aumenta anche y
x = fuel_consumption_array
y = emission_array

fig, ax = plt.subplots(figsize = (7, 7))

ax.scatter(x, y, s = 5, alpha = 0.5)

ax.set_xlabel("fuel consumption [l / 100km]")
ax.set_ylabel("emission")
ax.set_title("consumption")

ax.grid()
plt.show()
```

```python
dizionario_marche_occorrenze = {}
for marca in make_list:
    if marca not in dizionario_marche_occorrenze:
        dizionario_marche_occorrenze[marca] = 1
    else:
        dizionario_marche_occorrenze[marca] += 1
```

```python
lista_marche_occorrenze_ordinato = list(dizionario_marche_occorrenze.items())
```

```python
lista_marche_occorrenze_ordinato.sort()
```
