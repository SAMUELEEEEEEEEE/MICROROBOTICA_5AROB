```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("mnist_784_csv.csv", "r", encoding = "utf-8") as file:
    lines = file.readlines()
```

```python
numbers = []
images = []
for line in lines[1 : ]:
    number_to_represent = int(line.split(",")[784])
    numbers.append(number_to_represent)
    
    image = np.array([int(number) for number in line.split(",")])
    images.append(image)
    
```

```python
numbers = np.array(numbers)
images = np.array(images)
```

```python
fig, axs = plt.subplots(ncols = 10, nrows = 10, figsize = (20, 30))

for number in range(10):
    for index_photo in range(10):
        arraybool = numbers ==  number
        images_showable = images[arraybool]
        
        index = np.random.randint(0, len(images_showable))
        
        axs[number][index_photo].imshow(images_showable[index][ : -1].reshape((28, 28)), cmap = "Greys")
        axs[number][index_photo].axis("off")
plt.show()
```
