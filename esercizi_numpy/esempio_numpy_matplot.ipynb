```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size_x, size_y = 100, 100
image = np.zeros((size_x, size_y), dtype = np.int16)
```

```python
image[0 : 5, :] = 255
image[-5 : , :] = 255
image[:, 0 : 5] = 255
image[:, -5 : ] = 255
```

```python
plt.imshow(image)
```

```python
image[size_x // 2, size_y // 2] = 100
```

```python
plt.imshow(image, cmap = "gnuplot")
```

```python
image[size_x // 2 - 5 : size_x // 2 + 5 , size_y // 2 - 5: size_y // 2 + 5] = 100
```

```python
plt.imshow(image, cmap = "gnuplot")
```

```python
size_x, size_y = 100, 100
image = np.zeros((size_x, size_y), dtype = np.int16)
image[: , ] = np.arange(0, size_y)
```

```python
image
```

```python
plt.imshow(image, cmap = "Blues")
```

```python
image[size_x // 2: ] = 0
```

```python
plt.imshow(image, cmap = "Blues")
```

```python
size_x, size_y = 100, 100
image = np.zeros((size_x, size_y), dtype = np.int16)
```

```python
n_dots = 50
x = np.random.randint(0, size_x, n_dots)
y = np.random.randint(0, size_y, n_dots)
```

```python
image[x, y] = 500
```

```python
plt.imshow(image, cmap = "Blues")
```

```python
size_x, size_y = 100, 100
image = np.zeros((size_x, size_y), dtype = np.int16)
image[: , ] = np.arange(0, size_y)
image[(image % 2 == 1)] = 0
```

```python
plt.imshow(image, cmap = "Blues")
```
